## Limpieza de datos

Para limpiar los datos, se propone juntar todas las tablas cargadas por cada departamento y unirlas en una sola. Sin embargo, antes de eso se observó que el archivo "Colegios_Ciudad Capital.csv" no posee el departamento correcto y en la variable municipio posee el dato de zona, por lo tanto, este archivo requiere de una limpieza y transformación sencilla antes de juntar todos los datos de los departamentos en una misma tabla.

Se le asigna 

```python
dfs[4]["ZONA"] = dfs[4]["MUNICIPIO"] 
dfs[4]["MUNICIPIO"] = "GUATEMALA"
dfs[4]["DEPARTAMENTO"] = "GUATEMALA"
```

```python
dfs[4]
```

Ahora ya es posible unir todos los archivos en uno solo y comenzar la limpieza variable por variable.

```python
data = pd.concat(dfs, sort=False)
data
```

### Limpieza para variable: CÓDIGO

```python
data["CODIGO"] = data["CODIGO"].map(lambda x: int(str(x).replace("-","")))
data["CODIGO"]
```

Como se puede observar con esta simple transformación hemos transformado aquellos valores que causaban que el tipo de nuestra variable fuera objeto, ahora es un entero, además de que se han eliminado todos los guiones

### Limpieza para variable: DISTRITO

```python
data["DISTRITO"] = data["DISTRITO"].map(lambda x: np.nan if str(x)=='nan' else int(str(x).replace("-",""))).astype('Int64')
data["DISTRITO"]
```

Ahora hemos cambiado los valores de distrito al igual que el código eliminando los guiones, y convirtiendo los valores null en casillas identificables por pandas, es decir np.nan

### Limpieza para variable: DEPARTAMENTO
Para realizar la limpieza de esta variable debemos de revisar que esté bien escrito es por eso que vamos a realizar una tabla de frecuencia esperando tener 22 departamentos, pero antes es necesario revisar porque el valor de dtype es object

```python
data["DEPARTAMENTO"]
```

```python
table = data.groupby("DEPARTAMENTO").size().reset_index(name="Frecuencia")
table
```

Al revisar los datos se puede observar que el typo object viene de las distintas longitudes de los string generando que numpy lo entienda como object debido a que son strings de 32 bits y de 64 bits. Por lo que no hay que preocuparse, la variable esta limpia
